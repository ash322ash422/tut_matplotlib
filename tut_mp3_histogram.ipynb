# histogram
- Used for checking frequency/count of data

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Lets generate some data
np.random.seed(42) # Set seed for reproducibility
size = 100
mean = 40
std=20 # 68% of the value would be +/-20 around the mean
# Generate normally distributed values with mean=40 and std=20
random_normal = np.random.normal(loc=mean, scale=std, size=size) # age of people from  a sample of size

# Clip values to be within 20 and 150, then convert to integers
ages = np.clip(random_normal, 20, 150).astype(int) # LOOK AT  numpy tutorial on clip function
ages = np.sort(ages)

print(f"ages: \n{ages}")
```

```python
plt.hist(ages,
        bins = 4,
        edgecolor = 'black',
        color='skyblue'
)
plt.show()
```

### now I want to print the count, bins coordinates

```python
counts, bins, patches = plt.hist(ages,
        bins = 4,
        edgecolor = 'black',
        color='skyblue'
)

plt.show()

print(f"counts: {counts}") # This gives count in  each bin
print(f"bins: {bins}") # this gives the x-coordinate of each bin
```

# a simple application Check if dice is loaded

```python
# generate data for 6 sided dice
import random

s1 = [1] * 10
s2 = [2] * 10
s3 = [3] * 15  # introduce load
s4 = [4] * 10
s5 = [5] * 10
s6 = [6] * 10

dice_rolls = s1 + s2 + s3 + s4 + s5 + s6
random.shuffle(dice_rolls)
print(f"dice_rolls: {dice_rolls}")

counts, bins, patches =plt.hist(dice_rolls, 
         bins=6,
         color='lightgreen', 
         edgecolor='black')
plt.title("Histogram of Dice Rolls")
plt.xlabel("Dice Number")
plt.ylabel("Frequency")
plt.show()

print(f"counts: {counts}") # This gives count in  each bin
print(f"bins: {bins}") # this gives the x-coordinate of each bin
```

# Advanced (OPTIONAL)

```python
# Histogram of 200 Uniformly Distributed Integers (20–100)
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)

# Generate 200 random float numbers from uniform distribution, then round to integers
random_uniform_ints = np.random.uniform(20, 101, size=200).astype(int)
print(f"random_uniform_ints: \n{np.sort(random_uniform_ints)}")

# Plot the histogram
plt.figure(figsize=(10, 6))

# NOTE: It returns counts, bins, patches and also plots
plt.hist(random_uniform_ints,
         bins=8,
         edgecolor='black',
         color='skyblue')

plt.title('Histogram of 200 Uniformly Distributed Integers (20–100)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## observation about bin size
Here **bin_width = (max_value - min_value)/num_of_bins = (99-20)/8 = 9.8**.
- The bin starts from minimum value, which is 20. The 1st bin would be from 20 to (20+9.8). Then we keep adding 9.8 to each to get next bin point.
- The bins go from 20 -> 29.8 -> 39.6 -> 49.4 ... so on

Lets use above and increase the number of bins from 8 to 16. 

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)

# Generate 200 random float numbers from uniform distribution, then round to integers
random_uniform_ints = np.random.uniform(20, 101, size=200).astype(int)
print(f"random_uniform_ints: \n{np.sort(random_uniform_ints)}")

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(random_uniform_ints,
         bins=16,
         edgecolor='black',
         color='skyblue')
plt.title('Histogram of 200 Uniformly Distributed Integers (20–100)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Now, I want to see labels on each bin with the frequency plotted on top.
I use same code as above but this time I add logic for seeing the frequency.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42) # Set seed for reproducibility

# Generate 200 random float numbers from uniform distribution, then round to integers
random_uniform_ints = np.random.uniform(20, 101, size=200).astype(int)
print(f"random_uniform_ints: \n{np.sort(random_uniform_ints)}")

# Plot the histogram
plt.figure(figsize=(10, 6))
counts, bins, patches = plt.hist(random_uniform_ints,
                                 bins=8,
                                 edgecolor='black',
                                 color='skyblue')

# Add frequency labels to each bar
for count, patch in zip(counts, patches):
    # Get the center x position of the bar
    x = patch.get_x() + patch.get_width() / 2
    y = patch.get_height()
    plt.text(x, y + 1, str(int(count)), ha='center', va='bottom', fontsize=9)

# Add plot details
plt.title('Histogram of 200 Uniformly Distributed Integers (20–100)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Now I also want to see the numbers that are in each bin

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42) # Set seed for reproducibility

# Generate 200 random float numbers from uniform distribution, then round to integers
random_uniform_ints = np.random.uniform(20, 101, size=200).astype(int)
print(f"random_uniform_ints: \n{np.sort(random_uniform_ints)}")

# Plot the histogram
plt.figure(figsize=(10, 6))
counts, bins, patches = plt.hist(random_uniform_ints,
                                 bins=8,
                                 edgecolor='black',
                                 color='skyblue')

# Add frequency labels to each bar
for count, patch in zip(counts, patches):
    # Get the center x position of the bar
    x = patch.get_x() + patch.get_width() / 2
    y = patch.get_height()
    plt.text(x, y + 1, str(int(count)), ha='center', va='bottom', fontsize=9)

# Print bin ranges and counts
print("Bin Ranges and Counts:")
for i in range(len(counts)):
    print(f"Bin {i+1}: {bins[i]:.2f} to {bins[i+1]:.2f} => {int(counts[i])} values")

# Add plot details
plt.title('Histogram of 200 Uniformly Distributed Integers (20–100)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Histogram of 250 Gaussian Distributed Integers (Clipped 20–150)

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42) # Set seed for reproducibility

# Generate 250 normally distributed values with mean=85 and std=25
random_normal = np.random.normal(loc=85, scale=25, size=250) # size change to 1000, 5000

#OPTIONAL
mean_value = np.mean(random_normal) # Calculate the mean and verify that this is same as above
print(f" Mean(Before clip): { mean_value}")


# Clip values to be within 20 and 150, then convert to integers
random_normal_ints = np.clip(random_normal, 20, 150).astype(int) # LOOK AT  numpy tutorial on clip function
mean_value = np.mean(random_normal) # Calculate the mean and verify that this is same as above
print(f" Mean(After clip): { mean_value}")

print(f"size of random_normal_ints: {len(random_normal_ints)}")
print(f"random_normal_ints: \n{np.sort(random_normal_ints)}")


# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(random_normal_ints,
         bins=15,
         edgecolor='black',
         color='salmon')
plt.title('Histogram of 250 Gaussian Distributed Integers (Clipped 20–150)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### observation
- Its almost normal distributiuon, but not quite there because we are using small size. If yu increase the size to 1000 and then 5000 and then yu will see normal distribution
