# histogram
- Used for checking frequency or count of data

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# step1: Generate data
np.random.seed(42)

# mean = 65, std. dev. = 5, size = 100
# data = np.random.normal(65, 5, 100)  # heights in inches
data = np.array([51, 55, 55, 55, 56, 56, 57, 57, 57, 57, 58, 58, 59, 59, 59,
    59, 59, 60, 60, 60, 61, 61, 61, 61, 61, 61, 62, 62, 62, 62, 62, 62, 62,
    62, 62, 62, 63, 63, 63, 63, 63, 63, 63, 63, 63, 63, 63, 63, 64, 64, 64,
    64, 64, 64, 65, 65, 65, 65, 65, 65, 65, 65, 66, 66, 66, 66, 66, 66, 66,
    66, 66, 66, 66, 66, 66, 67, 67, 67, 68, 68, 68, 68, 69, 69, 69, 69, 69,
    69, 69, 70, 70, 70, 71, 72, 72, 72, 72, 72, 72, 74]
)
data = np.sort(data.astype(int))
print("data:\n", ", ".join(map(str, data)))
```

```python
# step2: plot the histogram with only 2 bins
bin_count = 2

fig, ax = plt.subplots(facecolor='black')
ax.set_facecolor('black')

counts, bins, patches = ax.hist(
    data,
    bins=bin_count,
    edgecolor='green',         # white edges for contrast
    color='gold'               # brighter color for visibility
)
ax.set_title('Histogram of Heights', color='white')
ax.set_xlabel('Height (inches)', color='white')
ax.set_ylabel('Frequency', color='white')
ax.tick_params(colors='white')
plt.show()

print(f"counts: {counts}")
print(f"bins  : {np.round(bins,1)}")
```

## observation about bin size
Here **bin_width = (max_value - min_value)/num_of_bins = (74-51)/2 = 11.5**.
- The bin starts from minimum value, which is 51, and then increase by 11.5 units. 
- The bins go from 51 -> 62.5 -> 74
- bin 51-62.5 contains 36 elements, bin 62.5-74 contains 64 elements

**Lets use above and increase the number of bins**

```python
# step3: plot the histogram with 2, 4, 10 bins
bin_counts = [2, 4, 10]

for bin_count in  bin_counts:
    fig, ax = plt.subplots(facecolor='black')
    ax.set_facecolor('black')
    
    counts, bins, patches = ax.hist(
        data,
        bins=bin_count,
        edgecolor='green',         # white edges for contrast
        color='gold'               # brighter color for visibility
    )
    ax.set_title(f'Histogram of Heights({bin_count})', color='white')
    ax.set_xlabel('Height (inches)', color='white')
    ax.set_ylabel('Frequency', color='white')
    ax.tick_params(colors='white')
    plt.show()
    
    print(f"counts: {counts}")
    print(f"bins  : {np.round(bins,1)}", "\n\n")
```

```python
# step3: Now increase the data to 10000 and bins to 100
np.random.seed(42)

# mean = 65, std. dev. = 5, size =1000
data = np.random.normal(65, 5, 10000)  # heights in inches
bin_count = 100

fig, ax = plt.subplots(facecolor='black')
ax.set_facecolor('black')

counts, bins, patches = ax.hist(
    data,
    bins=bin_count,
    edgecolor='green',         # white edges for contrast
    color='gold'               # brighter color for visibility
)
ax.set_title('Histogram of Heights', color='white')
ax.set_xlabel('Height (inches)', color='white')
ax.set_ylabel('Frequency', color='white')
ax.tick_params(colors='white')
plt.show()
```
