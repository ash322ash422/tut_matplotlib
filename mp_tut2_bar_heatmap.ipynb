## Heatmaps
Heatmaps are a great way to represent data where the colors represent the magnitude of a phenomenon. They are widely used for representing correlation matrices, where they can highlight relationships between variables, or for displaying geographical data density.

```python
import matplotlib.pyplot as plt

import numpy as np

# Generating data
data = np.random.rand(5, 5) # Random data for demonstration
print(data.shape)
print(data)
print("*************")

# Creating the heatmap
plt.imshow(data, cmap='hot', interpolation='nearest')

# Customization
plt.colorbar()

plt.title('Heatmap Example')
plt.show()
```

```python
# Lets generate a data that have strongly correlated features and also weakly correl features.
# Then we plot them
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

# Generating sample data with strong and weak correlations
np.random.seed(42)  # For reproducibility
size = 10  # Number of samples

# Strongly correlated data
x1 = np.random.rand(size) * 10  # Base feature
x2 = x1 + np.random.normal(0, 1, size)  # Strongly correlated with x1
x3 = x1 * 2 + np.random.normal(0, 2, size)  # Also strongly correlated with x1

# Weakly correlated data
x4 = np.random.rand(size) * 10  # Independent feature
x5 = np.random.rand(size) * 10  # Another independent feature

# Creating a DataFrame
df = pd.DataFrame({
    'Feat 1': x1,
    'Feat 2 (Strong Corr)': x2,
    'Feat 3 (Strong Corr)': x3,
    'Feat 4 (Weak Corr)': x4,
    'Feat 5 (Weak Corr)': x5
})
print(df)
print("**************")
# Computing the correlation matrix
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
# Creating the heatmap
plt.figure(figsize=(8, 6))

# Creating the heatmap
plt.imshow(correlation_matrix, cmap='hot', interpolation='nearest')

# Customization
plt.colorbar()

plt.title('Heatmap Example')
plt.show()
```

```python
# Same as above, but using seaborn. This looks better
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

# Generating sample data with strong and weak correlations
np.random.seed(42)  # For reproducibility
size = 10  # Number of samples

# Strongly correlated data
x1 = np.random.rand(size) * 10  # Base feature
x2 = x1 + np.random.normal(0, 1, size)  # Strongly correlated with x1
x3 = x1 * 2 + np.random.normal(0, 2, size)  # Also strongly correlated with x1

# Weakly correlated data
x4 = np.random.rand(size) * 10  # Independent feature
x5 = np.random.rand(size) * 10  # Another independent feature

# Creating a DataFrame
df = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2 (Strong Correlation)': x2,
    'Feature 3 (Strong Correlation)': x3,
    'Feature 4 (Weak Correlation)': x4,
    'Feature 5 (Weak Correlation)': x5
})

# Computing the correlation matrix
correlation_matrix = df.corr()

# Creating the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f")

# Customization
plt.title('Heatmap of Correlations')
plt.show()
```
