# scatterplot()

**Use scatterplots when:**

* You want to study relationship between two numerical variables
* You want to detect trends or patterns

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load dataset
df = sns.load_dataset("tips")

# Display first few rows
df.head()
```

### Here we use tips dataset
Columns and their meaning:

- total_bill: Total bill (in USD) for the meal, including food and drinks
- tip: Tip amount given by the customer
- sex: Gender of the person paying the bill — "Male" or "Female"
- smoker: Whether the customer is a smoker — "Yes" or "No"
- day: Day of the week — "Thur", "Fri", "Sat", or "Sun"
- time: Time of the meal — "Lunch" or "Dinner"
- size: Number of people at the table (party size)

```python
# Summary statistics
df.describe()
```

```python
# 1) Basic scatter plot

sns.scatterplot(data=df, x='total_bill', y='tip' )

plt.title("Total Bill vs Tip")
plt.show()
```


* Relationship between two numerical variables
* Each dot = one observation


```python
# 2) Adding categories: Now I want to seperate based on sex: male and female
sns.scatterplot(data=df, x='total_bill', y='tip',
                hue='sex' # This should be category
)

plt.show()
```

* `hue` separates data by category
* Automatically adds legend

```python
# 3) Now I want to seperate based on sex: male and female and also display the time category

sns.scatterplot(data=df, x='total_bill', y='tip',
                hue='sex',   # should be category
                style='time' # should be category
)

plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

* `style` changes marker shapes
* Useful when comparing multiple categories

```python
# 4) Scatterplot with Size

sns.scatterplot(
    x="total_bill",
    y="tip",
    size="size",
    sizes=(20, 200),
    data=df
)

plt.title("Total Bill vs Tip (Group Size)")
plt.show()
```

* Size shows a third numerical variable
* Larger group → larger dot


```python
# 5) Multiple Dimensions


sns.scatterplot(
    x="total_bill",
    y="tip",
    hue="day",
    size="size",
    sizes=(30, 300),
    data=df
)

plt.title("Total Bill vs Tip (Day & Group Size)")
plt.show()
```

> Scatterplots can visualize **3–4 dimensions at once**
