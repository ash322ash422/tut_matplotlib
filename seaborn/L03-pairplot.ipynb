# pairplot()
- Use pairplot when you want to explore relationships between multiple numerical features at once

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
iris = sns.load_dataset("iris")

# Display first few rows
iris.head()
```

```python
# 
sns.pairplot( data=iris )

plt.title("Iris Dataset - Pairwise Relationships")

plt.show()
```

```python

sns.pairplot(
    data=iris,
    hue="species"
)

plt.show()
```

```python
penguins = sns.load_dataset("penguins")

penguins.head()
```

```python
sns.pairplot(  data=penguins )

plt.show()
```

```python
sns.pairplot(
    data=penguins,
    hue="species"
)

plt.show()
```
