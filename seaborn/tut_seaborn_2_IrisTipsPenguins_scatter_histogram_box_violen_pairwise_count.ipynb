```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# Seaborn Scatter Plot Examples

```python
# Dataset 1: Iris
iris = sns.load_dataset("iris")

print(iris.head(5))

sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species")
plt.title("Iris Dataset - Sepal Length vs Width")
plt.show()
```

```python
# Dataset 2: Tips
tips = sns.load_dataset("tips")
print("+++++\n\n", tips.head(5))

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="time")

plt.title("Tips Dataset - Total Bill vs Tip")

plt.show()
```

```python
# Dataset 3: Penguins
penguins = sns.load_dataset("penguins")

print("+++++\n\n", penguins.head(5))

sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species", style="sex")

plt.title("Penguins Dataset - Flipper Length vs Body Mass")

plt.show()
```

# Seaborn Histogram Plot Examples

```python
# Dataset 1: Iris
sns.histplot(data=iris, x="petal_length", hue="species", kde=True, palette="Set2")
plt.title("Iris Dataset - Petal Length Distribution")
plt.show()
print("++++\n\n")

# Dataset 2: Tips
sns.histplot(data=tips, x="total_bill", hue="sex", kde=True, palette="pastel")
plt.title("Tips Dataset - Total Bill Distribution by Gender")
plt.show()
print("++++\n\n")

# Dataset 3: Penguins
sns.histplot(data=penguins, x="body_mass_g", hue="species", kde=True, palette="muted")
plt.title("Penguins Dataset - Body Mass Distribution")
plt.show()
```

```python
# Seaborn Box Plot Examples

# Dataset 1: Iris
sns.boxplot(data=iris, x="species", y="sepal_length", palette="Set3")
plt.title("Iris Dataset - Sepal Length by Species")
plt.show()
print("++++\n\n")

# Dataset 2: Tips
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex", palette="pastel")
plt.title("Tips Dataset - Total Bill by Day and Gender")
plt.show()
print("++++\n\n")

# Dataset 3: Penguins
sns.boxplot(data=penguins, x="species", y="body_mass_g", hue="sex", palette="muted")
plt.title("Penguins Dataset - Body Mass by Species and Sex")
plt.show()
```

# Seaborn Violin Plot Examples

```python
# Dataset 1: Iris
sns.violinplot(data=iris, x="species", y="sepal_length", palette="Set2")
plt.title("Iris Dataset - Sepal Length by Species (Violin Plot)")
plt.show()
print("++++\n\n")

# Dataset 2: Tips
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", split=True, palette="pastel")
plt.title("Tips Dataset - Total Bill by Day and Gender (Violin Plot)")
plt.show()
print("++++\n\n")

# Dataset 3: Penguins
sns.violinplot(data=penguins, x="species", y="body_mass_g", hue="sex", split=True, palette="muted")
plt.title("Penguins Dataset - Body Mass by Species and Sex (Violin Plot)")
plt.show()
```

```python
# Seaborn Pairwise Relationships (Pairplot Examples)

# Dataset 1: Iris
sns.pairplot(iris, hue="species", palette="Set2")
plt.suptitle("Iris Dataset - Pairwise Relationships", y=1.02)
plt.show()
print("++++\n\n")

# Dataset 2: Tips
sns.pairplot(tips, hue="sex", vars=["total_bill", "tip", "size"], palette="pastel")
plt.suptitle("Tips Dataset - Pairwise Relationships", y=1.02)
plt.show()
print("++++\n\n")

# Dataset 3: Penguins
sns.pairplot(penguins.dropna(), hue="species", palette="muted")
plt.suptitle("Penguins Dataset - Pairwise Relationships", y=1.02)
plt.show()
```

# Correlation Matrix with Seaborn (Updated)

```python
print("\n=== Correlation Matrix ===")

iris_numeric = iris.select_dtypes(include=['float64', 'int64'])  # Select only numeric columns

# Compute the correlation matrix
correlation_matrix_iris = iris_numeric.corr()
print("correlation_matrix_iris:\n", correlation_matrix_iris)

# Create a heatmap of the correlation matrix for Iris dataset
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix_iris, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Iris Dataset - Correlation Matrix")
plt.show()
print("++++\n\n")

tips_numeric = tips.select_dtypes(include=['float64', 'int64'])

# Create a heatmap of the correlation matrix for Tips dataset
correlation_matrix_tips = tips_numeric.corr()
print("correlation_matrix_tips:\n", correlation_matrix_tips)

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix_tips, annot=True, cmap="YlGnBu", fmt=".2f", linewidths=0.5)
plt.title("Tips Dataset - Correlation Matrix")
plt.show()
print("++++\n\n")

penguins_numeric = penguins.select_dtypes(include=['float64', 'int64'])

# Create a heatmap of the correlation matrix for Penguins dataset
correlation_matrix_penguins = penguins_numeric.corr()
print("correlation_matrix_penguins:\n", correlation_matrix_penguins)

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix_penguins, annot=True, cmap="viridis", fmt=".2f", linewidths=0.5)
plt.title("Penguins Dataset - Correlation Matrix")
plt.show()
```

#  Count Plot of Categorical Values

```python
# Count plot for categorical values (species column)
plt.figure(figsize=(6, 4))
sns.countplot(data=iris, x="species")
plt.title("Iris Dataset - Count of Species")
plt.show()
print("++++\n\n")

# Count plot for categorical values (day column) with hue based on sex
plt.figure(figsize=(6, 4))
sns.countplot(data=tips, x="day", hue="sex")
plt.title("Tips Dataset - Count of Days by Sex")
plt.show()
print("++++\n\n")

# Count plot for categorical values (species column) with hue based on sex
plt.figure(figsize=(6, 4))
sns.countplot(data=penguins, x="species", hue="sex")
plt.title("Penguins Dataset - Count of Species by Sex")
plt.show()
```
