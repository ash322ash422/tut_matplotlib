# stripplot


A strip plot displays individual data points for a numerical variable across categories.

- Strip plot shows every single observation.
- Do NOT use Strip Plot for very large datasets

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
tips = sns.load_dataset("tips")

# Display first few rows
tips.head()
```

```python
sns.stripplot(
    data=tips,
    x="day",
    y="total_bill"
)

plt.title("Strip Plot of Total Bill per Day")
plt.show()
```

Each dot = one customer

Vertical spread = distribution

```python
# Add jitter to prevent points overlap.

sns.stripplot(
    data=tips,
    x="day",
    y="total_bill",
    jitter=True
)

plt.title("Strip Plot with Jitter")
plt.show()
```

```python
sns.stripplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="time", # category
    jitter=True
)

plt.title("Total Bill per Day by time")
plt.show()
```
