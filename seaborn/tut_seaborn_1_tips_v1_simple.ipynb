# SEABORN tutorial via tips datset

- Seaborn = Easy, beautiful plots.
- Matplotlib = Full control & customization.
- We use both together to get the best of both worlds.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Basic scatter plot:
age    = [20,     42,    50,    33]
income = [30000, 45000, 90000, 65000]

sns.scatterplot(x=age, y=income) 

plt.show() # in some environment, this is needed
```

### Here we use tips dataset
Columns and their meaning:

- total_bill: Total bill (in USD) for the meal, including food and drinks
- tip: Tip amount given by the customer
- sex: Gender of the person paying the bill — "Male" or "Female"
- smoker: Whether the customer is a smoker — "Yes" or "No"
- day: Day of the week — "Thur", "Fri", "Sat", or "Sun"
- time: Time of the meal — "Lunch" or "Dinner"
- size: Number of people at the table (party size)

```python
# seaborn gives us the ability to dowload dataset from the web and experiment with that.
# So lets work on one of the downloaded dataset.

df = sns.load_dataset("tips")
print(df)
```

```python
# scatterplot()
```

```python
# 1) Basic scatter plot

sns.scatterplot(data=df, x='total_bill', y='tip' )

plt.show()
```

```python
# 2) Adding categories: Now I want to seperate based on sex: male and female
sns.scatterplot(data=df, x='total_bill', y='tip',
                hue='sex' # This should be category
)

plt.show()
```

```python
# 3) Now I want to seperate based on sex: male and female and also display the time category
sns.scatterplot(data=df, x='total_bill', y='tip',
                hue='sex',   # should be category
                style='time' # should be category
)

plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

# histplot()

```python
# lets look at distribution total_bill

sns.histplot(data=df,  x='total_bill',
    kde=True, # This plots the smooth curve
    color="skyblue"
)

plt.title("Histogram with KDE of Total Bill")

plt.show()
```

## observation:
* Most people in this dataset spent between $10–$20 on their meal.
* A few people spent a lot more (up to ~$50), which the KDE curve will show as a long tail to the right.

# boxplot()

```python
# 1) Lets do box plot of tip with categorized by day
sns.boxplot(data=df,
            x='day', # x should be category,
            y='tip'  # y should be numeric
) 

plt.show()

# The circles are outliers
```

```python
# 2) now we add sex 
sns.boxplot(data=df,
            x='day',  # x should be category,
            y='tip',  # y should be numeric
            hue='sex' # hue should be categorical
) 

plt.show()

# observation: males are paying higher tips than female
```

```python
#  3) now lets do smoker 

sns.boxplot(data=df,
            x='day',  # x should be category,
            y='tip',  # y should be numeric
            hue='smoker' # hue should be categorical
) 

plt.title("Tip Distribution by Day and smoking status")
plt.show()

# observation: people who are smoker are paying more tips on Thur and Fri. On Sun, non smoker pay more
```

```python
# lets see total_bill
sns.boxplot(data=df,
            x='day',       # x should be category,
            y='total_bill' # y should be numeric
) 

plt.show()

# obs: people spend more money on weekend
```

```python
# # lets see total_bill categorized by Sex
sns.boxplot(data = df, x='day', y='total_bill', hue = 'sex') # x should be category, y should be numeric
plt.show()

# obs: male spend more money than female on fri
```

```python
# lets see total_bill categorized by time
sns.boxplot(data=df, x='day', y='total_bill', hue = 'time') # x should be category, y should be numeric
plt.show()

# obs: not many people come for dinner on thur
# on sat and sun, people come for dinner mainly
# on fri people come for both lunch and dinner
```

# violinplot()

```python
# 1) violin plot of total_bill categorized by day
sns.violinplot(data=df,
               x='day', # should be categorical
               y='total_bill' # should be numerical
)
plt.show()

# observation: people are spending more on sat and sun
```

```python
# 2) Violin plot of total_bill categorized by day and sex. To show distribution and density
sns.violinplot(data=df,
               x="day", # should be categorical
               y="total_bill", # should be numerical
               hue="sex", # should be categorical
               split=True
)

plt.title("Violin Plot of Total Bill by Day and Gender")
plt.show()

# obs: male are spending same as female on Thur...
```

# swarmplot()
- Ideal when you have < 5000 data points 

```python
# 1) 
sns.swarmplot(data=df,
              x='time', # should be categorical
              y='tip' # should be numerical
)

plt.show()
```

### NOTE:
- Each round dots are data points.

```python
# 2) Now lets compare with violin plot:

sns.violinplot(data=df,
              x='time', # should be categorical
              y='tip' # should be numerical
)

plt.show()
```

```python
# 3) Add sex category
sns.swarmplot(data=df,
              x='time', # should be categorical
              y='tip',  # should be numerical
              hue='sex' # should be categorical
)

plt.show()
```

```python
# 4) Add sex category and seperate them. I do not want overlapping.

sns.swarmplot(data=df,
              x='time', # should be categorical
              y='tip',  # should be numerical
              hue='sex', # should be categorical
              dodge=True # seperation
)
plt.show()
```

```python
# 4) Swam and violin  together

sns.swarmplot(data=df,
              x='time',
              y='tip',
              color='k'
)

sns.violinplot(data=df,
               x='time',
               y='tip',
               inner=None
)

plt.show()
```

```python
# 5) (OPTIONAL) Plot all 
fig, axes = plt.subplots(2, 2, figsize=(15,12)) # plot 4 graphs

sns.swarmplot(data=df, x='time', y='tip', ax=axes[0,0])

sns.swarmplot(data=df, x='time', y='tip', hue='sex', ax=axes[0,1]) # overlapping on sex feature

sns.swarmplot(data=df, x='time', y='tip', hue='sex', dodge=True, ax=axes[1,0]) # not overlapping on sex feature

# combine swarm and violin plot
sns.swarmplot(data=df, x='time', y='tip', color='k', ax=axes[1,1])
sns.violinplot(data=df, x='time', y='tip', inner=None, ax=axes[1,1])

plt.show()
```

# pairplot()
- Pairwise relationships in dataset
- Allows to see if there is any relation between data

```python
# 1) 
sns.pairplot(df)
plt.show()
```

```python
# 2) 
sns.pairplot(df, 
             hue="sex" # add color to male/female
)
plt.suptitle("Pair Plot of Tips Dataset", y=1.02)
plt.show()
```

## observation:
### Diagonal Plots: These are histograms (or KDEs) of each individual numeric variable, split by sex.

- total_bill diagonal plot: Shows the distribution of total bills (Males may have a slightly wider range, suggesting they might pay for higher bills more often.)
- tip diagonal plot: Distribution of tip values. (Both sexes give tips in similar ranges, though you spot that male group tends to give slightly more.)
- size diagonal plot: Shows group sizes. (Most group sizes are 2–4, regardless of gender.)

### total_bill vs tip:

- Positive correlation: higher total bill → higher tip (as expected).

### total_bill vs size:
- Generally, larger parties lead to larger bills.
- Points cluster around party size 2–4.
- Sex coloring can reveal if, say, males tend to be in slightly larger groups.


# heatmap()

```python
# Correlation matrix

mat_corr = df.corr(numeric_only=True)

sns.heatmap(mat_corr, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")

plt.show()
```

# countplot()

```python
# 1) Count of categorical values: Count of Visitors by Day 

sns.countplot(data=df, 
              x="day", 
)

plt.title("Count of Visitors by Day")
plt.show()

# obs: Lot more customers on Sat and Sun
```

```python
# 2) Count of categorical values: Count of Visitors by Day and time

sns.countplot(data=df, 
              x="day", 
              hue="time"
)

plt.title("Count of Visitors by Day and time")
plt.show()
```

### Note:
- For Thur, almost no dinner
- For sat and Sun, the restaurant serves only dinner and no lunch
