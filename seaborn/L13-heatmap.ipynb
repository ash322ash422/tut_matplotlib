```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
tips = sns.load_dataset("tips")

# Display first few rows
tips.head()
```

```python
# Correlation matrix

mat_corr = tips.corr(numeric_only=True)

mat_corr
```

```python
sns.heatmap(mat_corr, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")

plt.show()
```

```python
# 
sns.pairplot( data=tips )

plt.title("Tips Dataset - Pairwise Relationships")

plt.show()
```

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
flights.describe()
```

```python
flights_pivot = flights.pivot(
    index="month",
    columns="year",
    values="passengers"
)

flights_pivot
```

Rows  months

Columns  years

Cell  number of passengers

```python
sns.heatmap(flights_pivot)

plt.title("Flight Passengers Heatmap")
plt.show()
```

```python
sns.heatmap(
    flights_pivot,
    annot=True,
    fmt="d",
)

plt.title("Heatmap with Passenger Counts")
plt.show()
```

```python
# Better Readability

plt.figure(figsize=(12,6))

sns.heatmap(
    flights_pivot,
    
    # annot=True,
    # fmt="d",
    
    cmap="coolwarm",
    linewidths=0.5
)

plt.title("Monthly Flight Passengers Over Years")
plt.show()
```
