```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
tips = sns.load_dataset("tips")

# Display first few rows
tips.head()
```

```python
sns.barplot(
    data=tips,
    x="day",
    y="total_bill"
)

plt.title("Average Total Bill per Day")
plt.xlabel("Day")
plt.ylabel("Average Total Bill ($)")
plt.show()
```

Seaborn automatically calculates mean

Error bars show confidence interval

```python
sns.barplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex"
)

plt.title("Average Total Bill per Day by Gender")
plt.show()
```

```python
# show the sum of total bill

sns.barplot(
    data=tips,
    x="day",
    y="total_bill",
    estimator=sum,
    errorbar=None
)

plt.title("Total Bill per Day (Sum)")
plt.show()
```

```python
# Horizontal bar plot: When categories names are too long or too many

sns.barplot(
    data=tips,
    y="day",
    x="total_bill"
)

plt.title("Average Total Bill per Day")
plt.show()
```
