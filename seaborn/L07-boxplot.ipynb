# boxplot()

A **boxplot (box-and-whisker plot)** summarizes the **distribution of numerical data** using **five key statistics**:

1. Minimum
2. First Quartile (Q1 â€“ 25%)
3. Median (Q2 â€“ 50%)
4. Third Quartile (Q3 â€“ 75%)
5. Maximum



* **Bottom of box** â†’ Q1 (25%)
* **Top of box** â†’ Q3 (75%)
* **Height of box** â†’ IQR = Q3 âˆ’ Q1


**Middle 50% of the data lives inside the box**

## Why do we need a Boxplot?

### 1 Shows distribution at a glance

* Where is the data centered?
* How spread out is it?
* Is it skewed?

 Much faster than inspecting raw numbers.

---

### 2ï¸ Identifies outliers clearly

* Points beyond whiskers = **outliers**
* Extremely important in **data cleaning** and **EDA**

---

### 3ï¸ Excellent for comparison

* Compare distributions **across categories**
* Much better than histograms when comparing groups


### ðŸ”¹ Median Line

* Line inside the box
* Shows the **central value**
* If not centered â†’ data is skewed


### ðŸ”¹ Outliers

* Points beyond whiskers
* Indicate unusual or extreme values

 
> Outliers are not errors â€” they are values worth investigating.



```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")

tips.sample(10)
```

```python
# 1) Lets do box plot of tip 
sns.boxplot(data=tips,
            y='tip'  # y should be numeric
) 

plt.show()

# The circles are outliers
```

```python
# 2) Lets do box plot of tip with categorized by day
sns.boxplot(data=tips,
            x='day', # x should be category,
            y='tip'  # y should be numeric
) 

plt.show()

# The circles are outliers
```

```python
# 3) now we add sex 

sns.boxplot(data=tips,
            x='day',  # x should be category,
            y='tip',  # y should be numeric
            hue='sex' # hue should be categorical
) 

plt.show()
```

# observation: 
- males are paying higher tips than female

```python
# lets see total_bill categorized by time
sns.boxplot(data=tips,
            x='day', # category
            y='total_bill', # numerical
            hue = 'time') 

plt.show()
```

# observation: 
- not many people come for dinner on thur
- on sat and sun, people come for dinner mainly
- on fri people come for both lunch and dinner
