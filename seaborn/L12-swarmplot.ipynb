# swarmplot()
- Ideal when you have small dataset: < 5000 data points 

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
tips = sns.load_dataset("tips")

# Display first few rows
tips.head()
```

```python
sns.swarmplot(
    data=tips,
    x="day",
    y="total_bill"
)

plt.title("Swarm Plot of Total Bill per Day")
plt.show()
```

```python
# 1)
sns.swarmplot(data=tips,
              x='time', # should be categorical
              y='tip' # should be numerical
)

plt.show()
```

```python
# 2) Now lets compare with violin plot:

sns.violinplot(data=tips,
              x='time', # should be categorical
              y='tip' # should be numerical
)

plt.show()
```

```python
# 3) Add sex category
sns.swarmplot(data=tips,
              x='time', # should be categorical
              y='tip',  # should be numerical
              hue='sex' # should be categorical
)

plt.show()
```

```python
# 4) Add sex category and seperate them. I do not want overlapping.

sns.swarmplot(data=tips,
              x='time', # should be categorical
              y='tip',  # should be numerical
              hue='sex', # should be categorical
              dodge=True # seperation
)
plt.show()
```

```python
# 4) Swam and violin  together

sns.swarmplot(data=tips,
              x='time',
              y='tip',
              color='k'
)

sns.violinplot(data=tips,
               x='time',
               y='tip',
               inner=None
)

plt.show()
```

# swarmplot + boxplot

```python
sns.boxplot(
    data=tips,
    x="day",
    y="total_bill",
    color="lightgray"
)

sns.swarmplot(
    data=tips,
    x="day",
    y="total_bill",
    color="black"
)

plt.title("Boxplot + Swarm Plot")
plt.show()
```
