# countplot()

A count plot shows the number of observations in each category of a categorical variable.


- It answers “How many times does each category appear?”

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
tips = sns.load_dataset("tips")

# Display first few rows
tips.head()
```

```python
# 1) Count of categorical values: Count of Visitors by Day 

sns.countplot(data=tips, 
              x="day", 
)

plt.title("Count of Visitors by Day")
plt.show()

```

# observation:
- Lot more customers on Sat and Sun
- Less customer on fri
- 

```python
# 2) Count of categorical values: Count of Visitors by Day and time

sns.countplot(data=tips, 
              x="day", 
              hue="time"
)

plt.title("Count of Visitors by Day and time")
plt.show()
```

### Observation:
- For Thur, almost no dinner
- For sat and Sun, the restaurant serves only dinner and no lunch
