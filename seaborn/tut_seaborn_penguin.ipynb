```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# seaborn gives us the ability to dowload dataset from the web and experiment with that.
# So lets work on one of the downloaded dataset.

df = sns.load_dataset("penguins")
print(df)
```

```python
# 1) Basic scatter plot

sns.scatterplot(data=df, x='bill_length_mm', y='bill_depth_mm' )

plt.show()
```

```python
# 2) Adding categories: Now I want to seperate based on species
sns.scatterplot(data=df, x='bill_length_mm', y='bill_depth_mm',
                hue='species' # This should be category
)

plt.show()
```

# histplot

```python
# lets look at distribution total_bill

sns.histplot(data=df,  x='bill_length_mm',
    kde=True, # This plots the smooth curve
    color="skyblue"
)

plt.title("Histogram with KDE")

plt.show()
```

# boxplot()

```python
# 1) Lets do box plot of tip with categorized by day
sns.boxplot(data=df,
            y='flipper_length_mm', # y is numeric,
           
) 

plt.show()

# No outliers
```

```python
# 1) Lets do box plot of tip with categorized by day
sns.boxplot(data=df,
            y='flipper_length_mm', # y is numeric,
            x="species"
           
) 

plt.show()

#  outliers for Adelie
# flipper length is higher for Gentoo
```

# violinplot()

```python
# 1) violin plot 
sns.violinplot(data=df,
               y='flipper_length_mm' # should be numerical
)
plt.show()

# observation: 
```

```python
# 1) violin plot
sns.violinplot(data=df,
               x='species', # should be categorical
               y='flipper_length_mm' # should be numerical
)
plt.show()

# observation: 
```

```python
# 1) violin plot
sns.violinplot(data=df,
               x='species', # should be categorical
               hue='island',
               y='flipper_length_mm', # should be numerical
               split=True
)
plt.show()

# observation: 
```

# swarmplot()
- Ideal when you have < 5000 data points 

```python
# 1) 
sns.swarmplot(data=df,
              x='species', # should be categorical
              y='flipper_length_mm',  # should be numerical
              size=4
)

plt.show()
```

```python
# 1) 
sns.swarmplot(data=df,
              x='species', # should be categorical
              y='bill_length_mm',  # should be numerical
              size=4,
              hue='sex', # should be categorical
              dodge=True # seperation
              
)

plt.show()
```

```python
# 4) Swam and violin  together

sns.swarmplot(data=df,
              x='species', # should be categorical
              y='flipper_length_mm',  # should be numerical
              size=4,
              color='k'
)

sns.violinplot(data=df,
              x='species', # should be categorical
              y='flipper_length_mm',  # should be numerical
               inner=None
)

plt.show()
```

```python
# pairplot()
```

```python
# 1)
sns.pairplot(df)
plt.show()
```

```python
# 2) 
sns.pairplot(df, 
             hue="species" # add color based on category
)
plt.suptitle("Pair Plot", y=1.02)
plt.show()
```

# heatmap()

```python
# Correlation matrix

mat_corr = df.corr(numeric_only=True)

sns.heatmap(mat_corr, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")

plt.show()
```

```python
# countplot()
```

```python
# 1) Count of categorical values: 

sns.countplot(data=df, 
              x="species", 
)

plt.title("Count plot")
plt.show()

# obs:
```

```python
# 1) Count of categorical values: 

sns.countplot(data=df, 
              x="species", 
              hue="island"
)

plt.title("Count plot")
plt.show()

# obs: Adelie is present on all 3 islands
# Chinstrap in present on Dream island alone
# Gentoo is present on Biscoe alone.
```

# kdeplot()

```python
sns.kdeplot(
    data=df,
    x="flipper_length_mm"
)

plt.show()
```

```python
sns.kdeplot(
    data=df,
    x="flipper_length_mm",
    hue="species",
    fill=True
)

plt.show()
```

```python
sns.kdeplot(
    data=df,
    x="body_mass_g",
    hue="species",
    fill=True
)

plt.show()
```

```python
sns.kdeplot(
    data=df,
    x="flipper_length_mm",
    y="body_mass_g",
    hue="species",
    fill=True
)

plt.show()
```

# regplot()

```python
sns.regplot(
    data=df,
    x="flipper_length_mm",
    y="body_mass_g"
)

plt.show()
```

```python
sns.regplot(
    data=df,
    x="flipper_length_mm",
    y="body_mass_g",
    ci=None # no confidence interval
)

plt.show()
```

```python
sns.regplot(
    data=df,
    x="flipper_length_mm",
    y="body_mass_g",
    scatter_kws={"s": 10, "alpha": 0.5},
    ci=None
)

plt.show()
```

```python
# Now lets increase the order of poly equation to 2

sns.regplot(
    data=df,
    x="flipper_length_mm",
    y="body_mass_g",
    order=2,
    ci=None
)

plt.show()
```
