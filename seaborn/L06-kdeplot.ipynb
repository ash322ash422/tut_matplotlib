# kdeplot()

Kernel Density Estimation (KDE) is a way to visualize the probability density of a continuous variable.

- It is a smooth curve and an alternative to a histogram
- Area under the curve = 1



### Avoid KDE when:
- Dataset is very small
- Data is discrete
- Distribution has extreme outliers

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")

# Display first few rows
iris.head()
```

```python

# simple density function
sns.kdeplot( data=iris, x='sepal_length')

plt.show()
```

# observation:
- The sepal length lies between 3.5 to 8.8.
- Most of sepal length is between 4.5 and 7

```python
# simple density function
sns.kdeplot(data=iris, x='sepal_length', fill=True)

plt.show()
```

```python
# simple density function
sns.kdeplot(data=iris, x='petal_length', fill=True,
           hue="species" # categorical
           )

plt.show()
```

# Observation:
- Setosa has petal length between 1 and 2.2 . So we can easily identfy this species by looking at petal length
- Versicolor and virginica has little overlapping.
- The amount of area under KDE is almost similar for 3 species. That means we have almost same number of setosa, versicolor, and virginica.
