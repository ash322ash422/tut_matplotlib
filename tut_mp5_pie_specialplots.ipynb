```python
# application of pie:
from matplotlib import pyplot as plt
import pandas as pd
df = pd.read_csv('iris.csv')


# Count number of samples per species
species_counts = df['target'].value_counts() # this is DF
print(species_counts)

# Extract data, labels, and colors
data = species_counts.values
data_labels = species_counts.index.tolist()
data_colors = ['lightcoral', 'lightskyblue', 'yellowgreen']
data_explode = [0.05] * len(data)  # explode all slices slightly

# Create pie chart
plt.pie(
    data,
    labels=data_labels,
    shadow=True,
    colors=data_colors,
    explode=data_explode,
    startangle=90,
    autopct='%1.1f%%'  # display fraction as percentage
)

plt.legend(title="Iris Species", loc="best", fancybox=True)
plt.axis('equal')     # plot pie chart as a circle
plt.tight_layout()
plt.title("Distribution of Iris Species")
plt.show()
```

# (SKIP) Basic triangulation

```python
from matplotlib import pyplot as plt
import matplotlib.tri as tri
import numpy as np

# rand_data = np.random.randn(50, 2)
rand_data = np.random.randint(-20, 21, size=(5, 2))
print(rand_data[:,0])
print("+++++++++++++++++++++++++++")
print(rand_data[:,1])

triangulation = tri.Triangulation(x=rand_data[:,0], y=rand_data[:,1])

plt.triplot(triangulation)

plt.show()
```

```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
warnings.filterwarnings(action="default")

x = [1, 2, 3]

y_1 = [50, 60, 70]
y_2 = [20, 30, 40]

with plt.xkcd():

    plt.plot(x, y_1, marker='x')
    plt.plot(x, y_2, marker='^')

    plt.xlim([0, len(x)+1])
    plt.ylim([0, max(y_1+y_2) + 10])
    plt.xlabel('x-axis label')
    plt.ylabel('y-axis label')
    plt.title('Simple line plot')
    plt.legend(['sample 1', 'sample2'], loc='upper left')

    plt.show()
```

```python
import numpy as np
import random
from matplotlib import pyplot as plt

data = np.random.normal(0, 20, 1000) 
bins = np.arange(-100, 100, 5) # fixed bin size

with plt.xkcd():
    plt.xlim([min(data)-5, max(data)+5])

    plt.hist(data, bins=bins, alpha=0.5)
    plt.title('Random Gaussian data (fixed bin size)')
    plt.xlabel('variable X (bin size = 5)')
    plt.ylabel('count')

    plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np

with plt.xkcd():
    X = np.random.random_integers(1,5,5) # 5 random integers within 1-5
    cols = ['b', 'g', 'r', 'y', 'm']
    plt.pie(X, colors=cols)
    plt.legend(X)

    plt.show()
```
