```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 250 random float numbers from uniform distribution, then round to integers
random_uniform_ints = np.random.uniform(20, 150, size=250).astype(int)
print(f"random_uniform_ints: \n{np.sort(random_uniform_ints)}")

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(random_uniform_ints,
         bins=15,
         edgecolor='black',
         color='skyblue')
plt.title('Histogram of 250 Uniformly Distributed Integers (20–150)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 250 normally distributed values with mean=85 and std=25
random_normal = np.random.normal(loc=85, scale=25, size=250)

# Clip values to be within 20 and 150, then convert to integers
random_normal_ints = np.clip(random_normal, 20, 150).astype(int)
print(f"random_uniform_ints: \n{np.sort(random_uniform_ints)}")


# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(random_normal_ints, bins=15, edgecolor='black', color='salmon')
plt.title('Histogram of 250 Gaussian Distributed Integers (Clipped 20–150)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Boxplot

```python
# Create a function to calculate all boxplot statistics
def calculate_box_stats(data):
    # Calculate the box plot statistics
    min_val = np.min(data)
    Q1 = np.percentile(data, 25)
    median = np.median(data)
    Q3 = np.percentile(data, 75)
    max_val = np.max(data)
    
    # Calculate the interquartile range (IQR)
    IQR = Q3 - Q1
    
    # Identify outliers (any data points outside 1.5 * IQR above Q3 or below Q1)
    outliers = data[(data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))]
    
    return min_val, Q1, median, Q3, max_val, outliers

# Print out the values
def print_box_stats(stats):
    min_val, Q1, median, Q3, max_val, outliers = stats
    print(f"Salary Stats:")
    print(f"  Minimum: {min_val:.2f}")
    print(f"  Q1: {Q1:.2f}")
    print(f"  Median: {median:.2f}")
    print(f"  Q3: {Q3:.2f}")
    print(f"  Maximum: {max_val:.2f}")
    print(f"  Outliers: {outliers}")
    print()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Simulating height data for male and female students
np.random.seed(42)
male_heights = np.random.normal(loc=175, scale=10, size=100)
female_heights = np.random.normal(loc=165, scale=8, size=100)

# Create a box plot
plt.boxplot([male_heights, female_heights], labels=["Male", "Female"], patch_artist=True, 
            boxprops=dict(facecolor='lightblue', color='black'), flierprops=dict(markerfacecolor='red', marker='o'))
plt.title("Height Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Height (cm)")
plt.grid(True)
plt.show()

#################################
print("Male stats:")
stats = calculate_box_stats(male_heights)
print_box_stats(stats)


print("Female stats:")
stats = calculate_box_stats(female_heights)
print_box_stats(stats)
```

```python
# Simulating temperature data for three cities
city1_temps = np.random.normal(loc=15, scale=5, size=30)
city2_temps = np.random.normal(loc=22, scale=6, size=30)
city3_temps = np.random.normal(loc=10, scale=4, size=30)

# Create a box plot
plt.boxplot([city1_temps, city2_temps, city3_temps], labels=["City 1", "City 2", "City 3"], patch_artist=True, 
            boxprops=dict(facecolor='lightgreen', color='black'), flierprops=dict(markerfacecolor='red', marker='o'))
plt.title("Monthly Temperature Distribution in Cities")
plt.xlabel("City")
plt.ylabel("Temperature (°C)")
plt.grid(True)
plt.show()

####################################################
print("City1 stats:")
stats = calculate_box_stats(city1_temps)
print_box_stats(stats)


print("City2 stats:")
stats = calculate_box_stats(city2_temps)
print_box_stats(stats)

print("City3 stats:")
stats = calculate_box_stats(city3_temps)
print_box_stats(stats)
```
