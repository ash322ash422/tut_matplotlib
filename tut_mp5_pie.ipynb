# Basic pie chart

```python
# The bare minimum of pie plot

from matplotlib import pyplot as plt
data = [10, 5, 20] # coffee , tea , shake drinkers

plt.pie(data)
plt.show()
```

```python
# Lets prettify above
# NOTE: uncomment each one below to see what each one does
from matplotlib import pyplot as plt
data = [10, 5, 20] # coffee , tea , shake drinkers
data_labels = ['coffee', 'tea', 'shakes']
data_colors = ('yellowgreen', 'lightskyblue', 'gold')
data_explode = (0, 0.1, 0.1)

plt.pie(
    data,
    labels=data_labels,
    shadow=True,
    colors=data_colors,
    explode=data_explode, # space between slices 
    startangle=90,    # rotate conter-clockwise by 90 degrees
    autopct='%1.1f%%',# display fraction as percentage
)
plt.legend(fancybox=True)
plt.axis('equal')     # plot pyplot as circle
plt.tight_layout()

plt.show()
```

```python
# application of pie:
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('iris.csv')
print(df.head(7))

# Count number of samples per species
species_counts = df['species'].value_counts() # this is DF
print(species_counts)

# Extract data, labels, and colors
data = species_counts.values
data_labels = species_counts.index.tolist()
# print(f"data_labels: {data_labels}")

data_colors = ['lightcoral', 'lightskyblue', 'yellowgreen']
data_explode = [0.05, 0.05, 0.05] # [0.05] * len(data_labels) explode all slices slightly

# Create pie chart
plt.pie(
    data,
    labels=data_labels,
    shadow=True,
    colors=data_colors,
    explode=data_explode,
    startangle=90,
    autopct='%1.1f%%'  # display fraction as percentage
)

plt.legend(title="Iris Species", loc="best", fancybox=True)
plt.axis('equal')     # plot pie chart as a circle
plt.tight_layout()
plt.title("Distribution of Iris Species")
plt.show()
```
