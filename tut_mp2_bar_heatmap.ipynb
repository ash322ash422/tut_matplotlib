## Bar

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array(["B", "Z", "D", "K"]) # person name
y = np.array([ 3,   4,   3,   7]) # height, number of children, number of houses

plt.bar(x, y, color="green")
plt.show()
```

```python
# different colors on  bars
x = np.array(["B", "Z", "D", "K"]) # person name
y = np.array([ 3,   4,   3,   7]) # height

# use a list of colors
colors = ["red", "blue", "green", "orange"]

plt.bar(x, y, color=colors)
plt.show()
```

```python
# draw horizonatal bars
# different colors on  bars
x = np.array(["B", "Z", "D", "K"]) # person name
y = np.array([ 3,   4,   3,   7]) # height

# use a list of colors
colors = ["red", "blue", "green", "orange"]

plt.barh(x, y, color=colors) # changed bar -> barh
plt.show()
```

```python
# now lets change the width
x = np.array(["B", "Z", "D", "K"]) # person name
y = np.array([ 3,   4,   3,   7]) # height

plt.bar(x, y, color="green", width = 0.1)
plt.show()
```

```python
# (optional)Advanced application of bar
x = np.array(["class9", "class10", "class11", "class12"]) # 4 classes in  chools

# male/female counts for two classes
male_counts   = np.array([3, 4, 3, 7])
female_counts = np.array([2, 3, 5, 8])

# Bar width and positions
bar_width = 0.3
x_pos = np.arange(len(x))

# Plotting grouped bars
plt.bar(x_pos - bar_width/2, male_counts, width=bar_width, label='Male', color='blue')
plt.bar(x_pos + bar_width/2, female_counts, width=bar_width, label='Female', color='orange')

# X-axis labels and legend
plt.xticks(x_pos, x)
plt.xlabel("Class")
plt.ylabel("Count")
plt.title("Count of People by Class")
plt.legend()
plt.tight_layout()
plt.show()
```

## Heatmap
- represent data where the colors represent the magnitude of a phenomenon.
-  They are widely used for representing correlation matrices, where they can highlight relationships between variables, or for displaying geographical data density.


```python
# generating 2D data
data = np.random.rand(5,5)
print(data.shape)
print(data)
print("*******************")

plt.imshow(data, cmap='hot') # small number are black and large numbers are white
# 'hot', 'viridis', 'plasma', 'inferno', 'magma', 'cividis', 'coolwarm'
plt.title("Heatmap example")
plt.show()
```

### application  of heatmap

```python
# step1 
# for reproducibilty
np.random.seed(42)

size = 10 # sample size
x1 = np.random.randint(20, 81, size)# age of a person between 20 and 80
print(f"x1: {x1}")

# strongly correlated x
x2 = x1 * 1000 + np.random.normal(0, 5000, size)  # income
x3 = x1 * 800 + np.random.normal(500,2000, size)  # balance in  bank account

# weakly correlated x
x4 = np.random.randint(2, 8, size)# number of  kids
x5 = np.random.randint(4, 50, size)# number of pushups


print(f"x2: {x2}")
print(f"x3: {x3}")
print(f"x4: {x4}")
print(f"x5: {x5}")
```

```python
# step2
# lets see if there is any correlation  between variables
plt.scatter(x1,x2, color='green')
plt.show()

plt.scatter(x1,x3, color='blue')
plt.show()

plt.scatter(x1,x4, color='red')
plt.show()

plt.scatter(x1,x5, color='magenta')
plt.show()

```

```python
# Aside: we can plot the pair plot using seaborn or pandas. Here I show using pandas. (Also see seaborn tutorial on how to perform pair plots)
import pandas as pd
from pandas.plotting import scatter_matrix

df = pd.DataFrame({
    'Age': x1,
    'Income': x2,
    'Bank Balance': x3,
    'Kids': x4,
    'push ups': x5
})

scatter_matrix(df, alpha=0.8, figsize=(10, 10), diagonal='hist', marker='o', grid=True)
plt.suptitle("Scatter Matrix of Sample Data", fontsize=16)
plt.show()
```

```python
# step3. Create correlation  matrix
# Combine into 2D array
data = np.vstack([x1, x2, x3, x4, x5])

# Calculate correlation matrix
corr_matrix = np.corrcoef(data)
print(corr_matrix)
```

```python
# step4 plot the heatmap
plt.imshow(corr_matrix, cmap='coolwarm') # small number are blue and large numbers are red
plt.show()
```

```python
# step5. Lets prettify it with labels, color bar etc
# Plot heatmap
fig, ax = plt.subplots(figsize=(6, 5))
im = ax.imshow(corr_matrix, cmap='coolwarm', vmin=-1, vmax=1)

# Add colorbar
cbar = plt.colorbar(im, ax=ax)
cbar.set_label('Correlation Coefficient')

# Set ticks and labels
labels = ['Age', 'Income', 'Bank Balance', 'Kids', 'push ups']
ax.set_xticks(np.arange(len(labels)))
ax.set_yticks(np.arange(len(labels)))
ax.set_xticklabels(labels, rotation=45, ha='right')
ax.set_yticklabels(labels)

# Add title
plt.title('Correlation Heatmap')

plt.show()
```

```python
# step6 Lets annotate each cell
# Plot heatmap
fig, ax = plt.subplots(figsize=(6, 5))
im = ax.imshow(corr_matrix, cmap='coolwarm', vmin=-1, vmax=1)

# Add colorbar
cbar = plt.colorbar(im, ax=ax)
cbar.set_label('Correlation Coefficient')

# Set ticks and labels
labels = ['Age', 'Income', 'Bank Balance', 'Kids', 'push ups']
ax.set_xticks(np.arange(len(labels)))
ax.set_yticks(np.arange(len(labels)))
ax.set_xticklabels(labels, rotation=45, ha='right')
ax.set_yticklabels(labels)

# Annotate each cell with the correlation value
for i in range(len(labels)):
    for j in range(len(labels)):
        text = ax.text(j, i, f"{corr_matrix[i, j]:.2f}",
                       ha="center", va="center", color="black")


# Add title
plt.title('Correlation Heatmap')

plt.show()
```
