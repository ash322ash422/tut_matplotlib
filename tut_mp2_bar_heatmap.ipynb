## Bar

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = ["B", "Z", "D", "K"] # person name
y = [ 3,   4,   3,   7] # height, number of children, number of houses

# or use numpy array
# x = np.array(["B", "Z", "D", "K"]) # person name
# y = np.array([ 3,   4,   3,   7]) # height

plt.bar(x, y, color="green")
plt.show()
```

```python
# different colors on  bars
x = ["B", "Z", "D", "K"] # person name
y = [ 3,   4,   3,   7] # height

# use a list of colors
colors = ["red", "blue", "green", "orange"]

plt.bar(x, y, color=colors)
plt.show()
```

```python
# draw horizonatal bars
# different colors on  bars
x = ["B", "Z", "D", "K"] # person name
y = [ 3,   4,   3,   7] # height

# use a list of colors
colors = ["red", "blue", "green", "orange"]

plt.barh(x, y, color=colors) # changed bar -> barh
plt.show()
```

```python
# now lets change the width
x = ["B", "Z", "D", "K"] # person name
y = [ 3,   4,   3,   7] # height

plt.bar(x, y, color="green", width = 0.1)
plt.show()
```

```python
# (optional)Advanced application of bar
x = np.array(["class9", "class10", "class11", "class12"]) # 4 classes in  chools

# male/female counts for two classes
male_counts   = np.array([3, 4, 3, 7])
female_counts = np.array([2, 3, 5, 8])

# Bar width and positions
bar_width = 0.3
x_pos = np.arange(len(x))

# Plotting grouped bars
plt.bar(x_pos - bar_width/2, male_counts, width=bar_width, label='Male', color='blue')
plt.bar(x_pos + bar_width/2, female_counts, width=bar_width, label='Female', color='orange')

# X-axis labels and legend
plt.xticks(x_pos, x)
plt.xlabel("Class")
plt.ylabel("Count")
plt.title("Count of People by Class")
plt.legend()
plt.tight_layout()
plt.show()
```

## Heatmap
- represent data where the colors represent the magnitude of a phenomenon.
-  They are widely used for representing correlation matrices, where they can highlight relationships between variables, or for displaying geographical data density.


```python
# generating 2D data
data = np.random.rand(5,5)
print(data.shape)
print(data)
print("*******************")

plt.imshow(data, cmap='coolwarm') # small number are black and large numbers are white
# 'hot', 'viridis', 'plasma', 'inferno', 'magma', 'cividis', 'coolwarm'
plt.title("Heatmap example")
plt.colorbar()
plt.show()
```
