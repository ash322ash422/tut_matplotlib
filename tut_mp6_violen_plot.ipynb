## Violin Plot
Violin Plot is used to visualize the distribution of numerical data of different variables. It is similar to Box Plot but with a rotated plot on each side, giving more information about the density estimate on the y-axis. The density is mirrored and flipped over, and the resulting shape is filled in, creating an image resembling a violin. Boxplot shows the outliers in the data lot clearly. 

Violin plots are used when you want to observe the distribution of numeric data, and are especially useful when you want to make a comparison of distributions between multiple groups. The peaks, valleys, and tails of each group’s density curve can be compared to see where groups are similar or different. Additional elements, like box plot quartiles, are often added to a violin plot to provide additional ways of comparing groups
Violin Plots hold more information than box plots, they are less popular, because their meaning can be harder to grasp for many readers not familiar with the violin plot representation.

The width of each curve corresponds with the approximate frequency of data points in each region.

A violin plot consists of four components. 

1. A white Centered Dot at the middle of the graph – The white dot point at the middle is the median of the distribution. 
2. A thin gray bar inside the plot – The bar in the plot represents the Quartile range of the distribution. 
3. A long thin line coming outside from the bar – The thin line represents the rest of the distribution which is calculated by the formulae Q1-1.5 IQR for the lower range and Q3+1.5 IQR for the upper range. The point lying beyond this line are considered as outliers.   
4. A line boundary separating the plot- A KDE plot is used for defining the boundary of the violin plot it represents the distribution of data points.

## bivariate analysis
In bivariate analysis, violin plots are utilized to examine the relationship between a continuous variable and a categorical variable. The categorical variable is represented on the x-axis, while the y-axis represents the values of the continuous variable.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# a simple violin plot
# Generate random data
np.random.seed(2)

# Simulate realistic student test scores (mean ~70, std dev ~10)
scores = np.random.normal(loc=70, scale=10, size=100) 
scores = np.round(scores,2)
# print(f"Student Scores(BEFORE):\n{scores}") 

scores = np.clip(scores, 0, 100)  # Ensure scores are between 0 and 100

print(f"Student Scores:\n{scores}")
print("**************")

# Create a violin plot
plt.figure(figsize=(6, 4))
plt.violinplot(scores,
               showmedians=True
)

# Add labels for context
# plt.title("Distribution of Student Test Scores")
# plt.ylabel("Score out of 100")
# plt.xticks([1], ["Test 1"])
# plt.grid(True)
# plt.tight_layout()

plt.show()
```

```python
data = df["sepal length (cm)"]

plt.violinplot(data,
               # showmeans=True,
               showmedians=True
)
plt.title("Violin Plot of Sepal Length")
plt.ylabel("Sepal Length (cm)")
plt.xticks([1], ['sepal length (cm)'])  # Set x-axis label

plt.grid(True)
plt.show()
```
